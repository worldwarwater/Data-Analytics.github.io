```python
# List to store book data
books = []

# Start from the first page
page_url = base_url + "index.html"

while page_url:
    # Request the page content
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all book containers
    for book in soup.find_all("article", class_="product_pod"):
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text
        availability = book.find("p", class_="instock availability").text.strip()
        books.append([title, price, availability])

    # Check if there's a "next" page
    next_button = soup.find("li", class_="next")
    if next_button:
        next_page = next_button.a["href"]
        page_url = base_url + next_page  # Update URL for the next page
    else:
        page_url = None  # Stop loop if no next page

# Save to CSV file
with open("fiction_books.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Price", "Availability"])
    writer.writerows(books)

print("Scraping completed! Data saved to 'fiction_books.csv'.")
```
